```python
import pandas as pd
tablita=pd.DataFrame(Estudiantes)
tablita
```

```python
tablita.describe()
```

```python
import seaborn as sns
sns.boxplot(tablita)
```

```python
import plotly.express as px
import pandas as pd

# Transform data to long format for plotly express
df_long = tablita.melt(var_name='Student', value_name='Grade')

fig = px.box(
    df_long,
    x='Student',
    y='Grade',
    points='all',  # This shows all data points
    title='Distribution of Student Grades',
    labels={'Grade': 'Grade', 'Student': 'Student'},
    template='plotly_white'
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

Estudiantes={"Sam":[3,4,5.9,8,3.3],
             "Ángel":[-1.5,0,2,5,10],
             "Rangel":[7.7,8.0,8.2,9,9],
             "Zarate":[0,10,8,3.5,0],
             "Ricardo":[3,6.2,7,2,8]}

tablita = pd.DataFrame(Estudiantes)

# Calcular el promedio de cada estudiante
promedios = tablita.mean()

# Crear un DataFrame para la gráfica de pastel
df_promedios = promedios.reset_index()
df_promedios.columns = ['Estudiante', 'Promedio']

# Crear la gráfica de pastel interactiva
fig = px.pie(
    df_promedios,
    values='Promedio',
    names='Estudiante',
    title='Contribución de promedios de calificaciones por estudiante',
    labels={'Promedio': 'Promedio', 'Estudiante': 'Estudiante'},
    template='plotly_white'
)

fig.show()
```
